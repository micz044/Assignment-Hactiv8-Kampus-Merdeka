```python
import pandas as pd
import numpy as np
import math
import statistics
import scipy.stats
import matplotlib.pyplot as plt
```

memanggil dan Menampilkan data yang ada didalam dataset

```python
# Konversi 'SALE PRICE' ke tipe data numerik
df['SALE PRICE']=df['SALE PRICE'].replace(' -  ', pd.NA)
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'], errors='coerce')
df['SALE PRICE']=df['SALE PRICE'].fillna(0)
df['SALE PRICE'] = df['SALE PRICE'].astype('int64')
```

pengecekan kolom yang ada di dataset untuk mengetahui apakah ada nilai null didalam kolom.

```python
# Count the number of 0 values in the 'SALE PRICE' column
sum = (df['SALE PRICE'] == 0).sum()
sum
```

```python
df.isnull().sum()
```

dilihat pada kode diatas, pada kolom/feature "SALE PRICE" terdapat nilai null sebanyak 14561. ini merupakan nilai yang cukup besar sehingga dapat memengaruhi hasil analisis kita nanti sehingga ini perlu di treatment.

```python
mean_price = df['SALE PRICE'].mean()
df['SALE PRICE'] = df['SALE PRICE'].replace(0, mean_price)
```

```python
# Membuat mapping antara nilai baru dan nilai asli borough
mapping_borough = {
    1: 'Manhattan',
    2: 'Bronx',
    3: 'Brooklyn',
    4: 'Queens',
    5: 'Staten Island'
}

# Mengganti nilai kolom 'BOROUGH' dengan nilai asli menggunakan mapping
df['BOROUGH'] = df['BOROUGH'].map(mapping_borough)
```

Seletah melalui Data Understanding, kami ingin menemukan insight yang ada didalam data ini. kami ingin menggunakan measure of central tendency seperti mean, median,modus,range, variance,standard Deviation,Probability distribution. kami ingin menggunakan mean untuk mengetahui rata rata tahun pembangunan gedung di masing-masing borough.

```python
rata_rata_per_borough = df.groupby('BOROUGH')['YEAR BUILT'].mean()
print("Rata-rata tahun pembangunan properti adalah:", rata_rata_per_borough)
```

Berdasarkan nilai rata rata diatas, dapat diketahui bahwa pada daerah Bronx, rata rata gedung yang dibangun pada tahun 1745, pada Brooklyn tahun 1767, dll.

kami juga ingin mengetahui pada tanggal bulan dan tahun berapa terjadi transaksi terbanyak dengan memanfaatkan modus dalam mencarinya.

```python
# Menghitung modus dari kolom 'SALE DATE'
modus_sale_date = df['SALE DATE'].mode()

# Menghitung frekuensi kemunculan setiap tanggal
frekuensi_tanggal = df['SALE DATE'].dt.date.value_counts()

# Menemukan tanggal yang menjadi modus
tanggal_modus = frekuensi_tanggal.idxmax()

# Menampilkan tanggal yang menjadi modus dan frekuensi kemunculannya
print("Tanggal Modus SALE DATE:", tanggal_modus)
print("Frekuensi Kemunculan:", frekuensi_tanggal.max())
```

Kemudian, saya ingin mengetahui berapa jumlah transaksi terbesar terjadi pada bulan dan tanggal berapa selama dari awal 2016 hingga akhir tahun 2016

```python
# Tentukan range (misalnya, Januari 2023 sampai Maret 2023)
start_date = pd.to_datetime('2016-01-01')
end_date = pd.to_datetime('2016-12-31')

# Mengubah kolom 'SALE DATE' ke tipe data datetime
df['SALE DATE'] = pd.to_datetime(df['SALE DATE'], errors='coerce')
# Filter dataset berdasarkan range tanggal
filtered_df = df[(df['SALE DATE'] >= start_date) & (df['SALE DATE'] <= end_date)]


# Periksa apakah ada data yang tersedia dalam rentang waktu
if not filtered_df.empty:
    # Menghitung frekuensi kemunculan setiap tanggal dalam range
    frekuensi_tanggal_range = filtered_df['SALE DATE'].dt.date.value_counts()

    # Menemukan tanggal yang menjadi modus dalam range
    tanggal_modus_range = frekuensi_tanggal_range.idxmax()

    # Menghitung total transaksi pada tanggal modus
    total_transaksi_modus = filtered_df[filtered_df['SALE DATE'].dt.date == tanggal_modus_range]['SALE PRICE'].count()

    # Menampilkan tanggal yang menjadi modus, frekuensi kemunculan, dan total transaksi pada tanggal modus
    print("Tanggal Modus SALE DATE (dalam range):", tanggal_modus_range)
    print("Frekuensi Kemunculan (dalam range):", frekuensi_tanggal_range.max())
    print("Total Transaksi pada Tanggal Modus:", total_transaksi_modus)
else:
    print("Tidak ada data yang tersedia dalam rentang waktu yang diberikan.")
```

diketahui dari data diatas, tingkat transaksi terbanyak terjadi pada 22 Desember 2016 dengan total transaksi terjadi sebanyak 527 kali.

```python
# Konversi 'SALE PRICE' ke tipe data numerik (jika mungkin)
df['SALE PRICE'] = pd.to_numeric(df['SALE PRICE'], errors='coerce')
sales_price = df['SALE PRICE']
# 4. Range
range_sales_price = np.max(sales_price) - np.min(sales_price)
print(f"Range Harga Penjualan pada sales price adalah: {range_sales_price}")
```

Range Harga Penjualan: Nilai ini menunjukkan selisih antara harga penjualan tertinggi dan terendah. Semakin besar nilai range, semakin besar variasi harga penjualan.

```python
# 5. Variance
variance_sales_price = np.var(sales_price)
print(f"Varians Harga Penjualan pada sales price adalah: {variance_sales_price:.2f}")
```

Varians Harga Penjualan: Varians mengukur seberapa jauh setiap nilai harga penjualan dari rata-rata. Varians yang tinggi menunjukkan bahwa harga penjualan tersebar lebih jauh dari rata-rata.

```python
# 6. Standard Deviation
std_dev_sales_price = np.std(sales_price)
print(f"Deviasi Standar Harga Penjualan pada sales price adalah: {std_dev_sales_price:.2f}")
```

Deviasi Standar Harga Penjualan: Deviasi standar adalah akar kuadrat dari varians. Nilai ini memberikan gambaran yang lebih mudah diinterpretasikan tentang sebaran harga penjualan. Semakin tinggi deviasi standar, semakin besar variasi harga penjualan.
